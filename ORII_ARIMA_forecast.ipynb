```python
df = pd.read_csv('/content/out.csv')
df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# switch here for quantity / income prediction
train = df[df.index <= 20]['Quantity']
test = df[df.index >= 20]['Quantity']
# train = df[df.index <= 20]['Income']
# test = df[df.index >= 20]['Income']
```

```python
from statsmodels.tsa.arima.model import ARIMA
y = train
ARIMAmodel = ARIMA(y, order = (3, 2, 3))
ARIMAmodel = ARIMAmodel.fit()

y_pred = ARIMAmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = ARIMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
```

```python
plt.plot(train, color = "black", label = "Training")
plt.plot(test, color = "red", label = "Testing")
plt.plot(y_pred_out, color='Blue', label = 'ARIMA Predictions')
plt.ylabel('Quantity')
plt.xlabel('Number of Days')
plt.rcParams['figure.figsize']=(10,5)
plt.title("ARIMA Prediction for ORII Dataset in NA")
plt.legend()
plt.show()
```
