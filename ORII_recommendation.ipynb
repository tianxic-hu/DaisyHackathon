```python
import collections
from pandas.core.common import Collection
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from gensim.models import Word2Vec
from sklearn.cluster import AgglomerativeClustering

# list of retail items
items = unique_des

# get the last two words of each description
descriptions = [item.split() for item in items]
last_two_words = [" ".join(description) for description in descriptions]

# convert items to a tf-idf matrix
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(last_two_words)

# perform k-means clustering
kmeans = KMeans(n_clusters=8, random_state=8).fit(X)

# alternatively, can try agglomerative hierarchical clustering
# ahc = AgglomerativeClustering(n_clusters=10)
# ahc.fit(X.toarray())

# print the cluster assignments for each item
clusters = {0: [], 1: [], 2: [], 3: [], 4: [], 5: [], 6: [], 7: []}
for i, item in enumerate(items):
    # print(f"Item: {item}, Cluster: {kmeans.labels_[i]}")
    clusters[kmeans.labels_[i]].append(item)
```

```python
# print 5 example items in each cluster 
for key, val in clusters.items():
  print(f"Cluster {key}: {val[5:10]}")
```

```python
# Try some new items! 
new_item = "RED POLKADOT CUPS"
# Convert the new item's description into a word vector
new_vector = vectorizer.transform([new_item])
# Use the predict() method to classify the new item
cluster_label = kmeans.predict(new_vector.toarray())
# Print the cluster label
print(f"The new item '{new_item}' belongs to cluster {cluster_label[0]}")
# Cluster 4: ['BLUE POLKADOT CUP', 'PINK  POLKADOT PLATE ', 'BLUE POLKADOT PLATE ', 'PARTY PIZZA DISH PINK POLKADOT', 'PARTY PIZZA DISH BLUE POLKADOT']
# So it indeed should be in cluster 4
```

```python
utility_matrix = df.pivot_table(values='Quantity', index='Customer ID', columns='ProductClass', fill_value=0)
utility_matrix.to_csv('utility_matrix.csv')
utility_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
customer_similarity = cosine_similarity(utility_matrix)
```

```python
# Recommending products for current users 
targets = [i for i in range(utility_matrix.shape[0])]
for target in targets:
    # Find the most similar customers
    similar = customer_similarity[target]

    # Sort the similar customers by similarity
    similar = similar.argsort()[::-1]

    # Get the products purchased by the similar customers, but not by the target customer
    products_to_recommend = []
    for similar_customer in similar:
        products = utility_matrix.columns[(utility_matrix.iloc[similar_customer] > 0) & (utility_matrix.iloc[target] == 0)]
        products_to_recommend.extend(products)

    # Print recommended products
    print(f"Recommended products for {target}th user: {products_to_recommend[:1]}")
```
